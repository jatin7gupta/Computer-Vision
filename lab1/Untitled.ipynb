## Read image file and show 

```python
def read_image_grey(path):
    return cv2.imread(path, 0)
    
def show_image(name, img):
    cv2.imshow(name, img)
    cv2.waitKey(0)

    # destroy all windows
    cv2.destroyAllWindows()
    
path = 'COMP9517_20T2_Lab1_Image/cat.png'
img = read_image_grey(path)
show_image(path, img)
```
